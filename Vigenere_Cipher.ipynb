## Implementation in Python

### Encrypy function

```python
# Vigènere cipher. Function to encrypt a given message using a given key and grouping by 't' letters
def encrypt(message, key, t):
  alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
  table = {}
  for i, a in enumerate(alphabet):
    for j, b in enumerate(alphabet):
      table[(a, b)] = alphabet[(i + j) % len(alphabet)]
  ciphertext = ''
  message = message.replace(' ', '')
  for index, letter in enumerate(message):
    if index % t == 0:
      ciphertext += ' '
    ciphertext += table[(letter, key[index % len(key)])] 
  return ciphertext.lstrip(' ')
```

### Decrypt function

```python
# Function to decrypt a given covered message using the Vigènere cipher
def decrypt(ciphertext, key, t):
  alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
  table = {}
  for i, a in enumerate(alphabet):
    for j, b in enumerate(alphabet):
      table[(a, b)] = alphabet[(i - j) % len(alphabet)]
  message = ''
  ciphertext = ciphertext.replace(' ', '')
  for index, letter in enumerate(ciphertext):
    if index % t == 0:
      message += ' '
    message += table[(letter, key[index % len(key)])] 
  return message.strip()
```

### Caesar function


```python
# Vigènere Cipher 
def vigenere(message, key, t, decrypt_mode=True):
  if decrypt_mode:
    return decrypt(message, key, t)
  else:
    return encrypt(message, key, t)
```

## Example

```python
# Define parameter 't' to group letters
t = 5
print("Size of string (t): \n", t)

# Define key to use in the algorithm
key = 'CRYPTOGRAPHY'
print("key: \n", key)

# Define the message to encrypt using this algorithm
message = 'I WILL ENCRYPT THIS MESSAGE TO COVER IMPORTANT INFORMATION' 
print("Message: \n", message)

# Cipher the message using the Vigènere Cipher 
ciphertext = vigenere(message, key, t, decrypt_mode=False)
print("Cipher text: \n", ciphertext)

# Recover the original message using the Vigènere Cipher  (undebugged message)
recovered_message = vigenere(ciphertext, key, t, decrypt_mode=True)
print("Recovered message (undebugged): \n", recovered_message)
```
